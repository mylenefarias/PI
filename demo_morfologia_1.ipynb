# Exemplos Morfologia

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('j.png',0)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.axis("off")
plt.show()
```

## Erosão

```python
kernel = np.ones((5,5),np.uint8)
erosao = cv2.erode(img,kernel,iterations = 1)
plt.imshow(cv2.cvtColor(erosao, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.axis("off")
plt.show()
```

```python
## Dilatação
```

```python
dilatacao = cv2.dilate(img,kernel,iterations = 1)
plt.imshow(cv2.cvtColor(dilatacao, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.axis("off")
plt.show()
```

```python
## Abertura
```

```python
abertura = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
plt.imshow(cv2.cvtColor(abertura, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.axis("off")
plt.show()
```

```python
## Fechamento
```

```python
fechamento = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
plt.imshow(cv2.cvtColor(fechamento, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.axis("off")
plt.show()
```

```python
## Gradiente
```

```python
gradiente = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
plt.imshow(cv2.cvtColor(gradiente, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.axis("off")
plt.show()
```

```python
## Top Hat
```

```python
kernel9 = np.ones((9,9),np.uint8)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel9)
plt.imshow(cv2.cvtColor(tophat, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.axis("off")
plt.show()
```
