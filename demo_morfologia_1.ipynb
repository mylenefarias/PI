# Exemplos Morfologia

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
# read as gray
img = cv2.imread('j.png',0)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.IMREAD_GRAYSCALE))
plt.axis("off")
plt.show()
```

## Erosão

```python
kernel = np.ones((5,5),np.uint8)
erosao = cv2.erode(img,kernel,iterations = 1)
plt.imshow(cv2.cvtColor(erosao, cv2.IMREAD_GRAYSCALE))
plt.axis("off")
plt.show()
```

```python
## Dilatação
```

```python
dilatacao = cv2.dilate(img,kernel,iterations = 1)
plt.imshow(cv2.cvtColor(dilatacao, cv2.IMREAD_GRAYSCALE))
plt.axis("off")
plt.show()
```

```python
# dilatação com uma outra imagem
img2 = cv2.imread('broken_text.png',0)
plt.imshow(cv2.cvtColor(img2, cv2.IMREAD_GRAYSCALE))
plt.axis("off")
plt.show()
```

```python
## Abertura
```

```python
# dilatação com uma outra imagem
img3 = cv2.imread('frame.png',0)
plt.imshow(cv2.cvtColor(img3, cv2.IMREAD_GRAYSCALE))
plt.axis("off")
plt.show()
```

```python
kernel = np.ones((20,20),np.uint8)
abertura = cv2.morphologyEx(img3, cv2.MORPH_OPEN, kernel)
plt.imshow(cv2.cvtColor(abertura, cv2.IMREAD_GRAYSCALE))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.axis("off")
plt.show()
```

```python
## Fechamento
```

```python
fechamento = cv2.morphologyEx(img3, cv2.MORPH_CLOSE, kernel)
plt.imshow(cv2.cvtColor(fechamento, cv2.IMREAD_GRAYSCALE))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.axis("off")
plt.show()
```

```python
## fechamento da abertura
fechamento2 = cv2.morphologyEx(abertura, cv2.MORPH_CLOSE, kernel)
plt.imshow(cv2.cvtColor(fechamento2, cv2.IMREAD_GRAYSCALE))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.axis("off")
plt.show()
```
